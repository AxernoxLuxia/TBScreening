```python
import shutil
import pandas as pd
import os

def move_files(source_dir, destination_dir, file_list):
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    for file_name in file_list:
        # Full file path
        source_file = os.path.join(source_dir, file_name)
        destination_file = os.path.join(destination_dir, file_name)

        try:
            # Move the file
            shutil.move(source_file, destination_file)
            print(f"Moved: {file_name}")
        except FileNotFoundError:
            print(f"File not found: {file_name}")
        except Exception as e:
            print(f"Error moving {file_name}: {e}")

source_dir = '/home/axernox/Resnet50/tbx11k-simplified1/traincl/healthy'
destination_dir = '/home/axernox/Resnet50/tbx11k-simplified/traincl/healthy'
csv_file = '/home/axernox/Resnet50/tbx11k-simplified/data1.csv'
df = pd.read_csv(csv_file)
file_list = []
for index, row in df.iterrows():
    im_type = row['image_type']
    im_name = row['fname']
    im_source = row['source']

    if im_type == 'healthy' and im_source == 'train':
        file_list.append(im_name)

print(file_list)
move_files(source_dir, destination_dir, file_list)
#move_files(source_dir, destination_dir, fl2)
```

```python
import shutil
import pandas as pd
import os

def copy_files(source_dir, destination_dir, file_list):
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    for file_name in file_list:
        
        source_file = os.path.join(source_dir, file_name)
        destination_file = os.path.join(destination_dir, file_name)

        try:
            shutil.copy2(source_file, destination_file)
            print(f"Copied: {file_name}")
        except FileNotFoundError:
            print(f"File not found: {file_name}")
        except Exception as e:
            print(f"Error copying {file_name}: {e}")

source_dir = '/home/axernox/Resnet50/tbx11k-simplified/valcl/tb'
destination_dir = '/home/axernox/Resnet50/tbx11k-simplified1/traincl/tb'
csv_file = '/home/axernox/Resnet50/tbx11k-simplified/data1.csv'
df = pd.read_csv(csv_file)
file_list = []
for index, row in df.iterrows():
    im_type = row['image_type']
    im_name = row['fname']
    im_source = row['source']

    if im_type == 'tb' and im_source == 'train':
        file_list.append(im_name)

print(file_list)
copy_files(source_dir, destination_dir, file_list)
#move_files(source_dir, destination_dir, fl2)
```

```python
import os
import pandas as pd

def delete_files_based_on_csv(directory, csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)
    
    # Counter for skipped files
    skipped_count = 0
    
    # Loop through each row in the DataFrame
    for index, row in df.iterrows():
        file_name = row['fname']  # Assuming the file names are in a 'file_name' column
        del_val = row['del_val']  # Check the 'del_val' column

        file_path = os.path.join(directory, file_name)
        
        if del_val == 'delete':
            try:
                if os.path.exists(file_path):
                    os.remove(file_path)
                    print(f"Deleted: {file_name}")
                else:
                    print(f"File not found: {file_name}")
            except Exception as e:
                print(f"Error deleting {file_name}: {e}")
        else:
            skipped_count += 1
            print(f"Skipped: {file_name}")

    print(f"\nTotal files skipped: {skipped_count}")

# Example usage
directory = '/home/axernox/Resnet50/tbx11k-simplified1/testcl/sick_no_tb'
csv_file = '/home/axernox/Resnet50/tbx11k-simplified1/data1.csv'

delete_files_based_on_csv(directory, csv_file)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Flatten, GlobalAveragePooling2D
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_height,img_width=256,256
batch_size=8
directory = '/home/axernox/Resnet50/tbx11k-simplified_can/traincan'
train = keras.utils.image_dataset_from_directory(
    directory,
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=8,
    image_size=(img_height,img_width),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True,
)
```

```python
directoryval = '/home/axernox/Resnet50/tbx11k-simplified_can/valcan'
val = keras.utils.image_dataset_from_directory(
    directoryval,
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=8,
    image_size=(img_height,img_width),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True,
)
```

```python
directorytest = '/home/axernox/Resnet50/tbx11k-simplified_can/testcan'
test = keras.utils.image_dataset_from_directory(
    directorytest,
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=8,
    image_size=(img_height,img_width),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True,
)
```

```python
datagen = ImageDataGenerator()

train_generator = datagen.flow_from_directory(
    '/home/axernox/Resnet50/tbx11k-simplified/train', 
    target_size=(512, 512), 
    batch_size=8,
    class_mode='categorical' 
)

# Print class indices (this shows the mapping between folder names and class labels)
print(train_generator.class_indices)

images, labels = next(train_generator)

# Print the shape of the images and labels
print("Images shape:", images.shape)  # Should be (batch_size, 512, 512, 3)
print("Labels shape:", labels.shape)  # If 'categorical', it will be (batch_size, num_classes)

# Print the first few labels
print("First 5 labels:", labels[:5])
```

```python
pretrained_model= tf.keras.applications.ResNet50(include_top=False,
                   input_shape=(256,256,3),
                   pooling='avg',classes=3,
                   weights=None)
for layer in pretrained_model.layers:
        layer.trainable=False
for layer in pretrained_model.layers[-10:]:
    layer.trainable = True

x = pretrained_model.output
x = Dense(512, activation='relu')(x)
output = Dense(3, activation='softmax')(x)      

resnet_model = Model(inputs=pretrained_model.input, outputs=output)
```

```python
resnet_model.summary()
```

```python
resnet_model.compile(optimizer=Adam(learning_rate=0.0001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = resnet_model.fit(
  train,
  validation_data=val,
  epochs=10
)
```

```python
train_loss_res, train_acc_res = resnet_model.evaluate(train)
```
